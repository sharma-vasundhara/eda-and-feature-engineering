### Finding Missing Values

```python
df.isnull().sum() # checking for missing values
```

```python
# finding features (columns) that have missing values
[features for features in df.columns if df[features].isnull().sum() > 1] 
```

```python
# creating heatmap for missing values in the dataset
plt.rcParams['figure.figsize'] = (12, 6) # setting size of plots to 12 x 6
sns.heatmap(df.isnull(), yticklabels = False, cbar = False, cmap = 'viridis')
```

As the dataset consists of 9551 rows and 21 columns, 8 missing values in `Cuisines` column is not visible on the heatmap

### Merge with Country Code
#### Merging the dataset with Country Code dataset to get country names

```python
# reading the file
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head() # checking the contents
```

```python
# merging the transaction dataset with country code dataset (left join)
final_df = pd.merge(df, df_country, on = 'Country Code', how = 'left')
final_df.head(3) # checking the contents of joined dataframe
```

```python
final_df.dtypes # checking column types
```

```python
country_names = final_df.Country.value_counts().index # country names
```

```python
country_values = final_df.Country.value_counts().values # count for each country
```

### Plot Pie Chart For Top 4 Countries

```python
plt.pie(x = country_values[:3], labels = country_names[:3], autopct = '%1.2f%%')
```

94.39% of the transactions are from India, followed by United States with 4.73%, and United Kingdom with 0.87%

### Relationships

```python
# diving into ratings
# grouping the data on the basis of rating columns
df_rating = final_df.groupby(['Aggregate rating', 
                              'Rating color', 
                              'Rating text']).size().reset_index().rename(columns = {0: 'Rating Count'})
df_rating
```

#### Observations

| Rating    | Rating Color | Rating Text |
|-----------|--------------|-------------|
| 4.5 - 4.9 | Dark Green   | Excellent   |
| 4.0 - 4.4 | Green        | Very Good   |
| 3.5 - 3.9 | Yellow       | Good        |
| 2.5 - 3.4 | Orange       | Average     | 
| 0 - 2.4   | Red          | Poor        |

```python
# plotting bar graph to check frequency of each rating
sns.barplot(data = df_rating, x = 'Aggregate rating', y = 'Rating Count')
```

```python
# coloring the bar graph on the basis of Rating Color in the dataframe
sns.barplot(data = df_rating, 
            x = 'Aggregate rating', 
            y = 'Rating Count', 
            hue = 'Rating color',
            palette = ['lightgrey', 'red', 'orange', 'yellow', 'limegreen', 'darkgreen'],
            dodge = False)
```

1. Not rated count is very high
2. Maximum number of ratings are between 2.5 and 3.9

```python
# plotting bar graph to check the frequency of each color in df_rating
sns.countplot(x = 'Rating color', 
              data = df_rating, 
              palette = ['lightgrey', 'red', 'orange', 'yellow', 'limegreen', 'darkgreen'])
```

```python
# countries that have given 0 ratigs
final_df[final_df['Aggregate rating'] == 0.0].groupby('Country').size().reset_index()
```

Maximum number of 0.0 ratings are from Indian customers

**Countries & Currencies**

```python
# checking countries and currencies
final_df.groupby(['Country', 'Currency']).size().reset_index()
```

**Which countries have online delivery?**

```python
# checking which countries have online delivery available
final_df.groupby(['Country', 'Has Online delivery']).size().reset_index()
```

### Plot Pie Chart For Top 4 Cities

```python
# checking top 4 cities with maximum transactions
df_cities = final_df[['City']].groupby('City').size().reset_index().rename(columns = {0: 'Count'}).sort_values('Count', ascending = False)
plt.pie(x = df_cities['Count'][:4], labels = df_cities['City'][:4], autopct = '%1.2f%%')
```
